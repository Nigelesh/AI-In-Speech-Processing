```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```python
audio_file_path = "C:/Users/nigel/Downloads/Nigelesh.wav"
audio_signal, sampling_rate = librosa.load(audio_file_path)

plt.figure(figsize=(12, 4))
librosa.display.waveshow(audio_signal, sr=sampling_rate)
plt.title("Waveform of the Audio File")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

## Q2

```python
signal_duration = len(audio_signal) / sampling_rate
magnitude_range = max(abs(audio_signal)) - min(abs(audio_signal))

print(f"Signal Duration: {signal_duration} seconds")
print(f"Magnitude Range: {magnitude_range}")
```

## Q3

```python
from IPython.display import Audio

start_time_seg = 0.5
end_time_seg = 1.5

segment_to_play = audio_signal[int(start_time_seg * sampling_rate):int(end_time_seg * sampling_rate)]

Audio(segment_to_play, rate=sampling_rate)
```

## Q4

```python
from IPython.display import Audio, display

seg1_start = 0.0
seg1_end = 1.0

seg2_start = 0.9
seg2_end = 1.6

abrupt_seg_start = 1.1
abrupt_seg_end = 1.9

segment_1_audio = audio_signal[int(seg1_start * sampling_rate):int(seg1_end * sampling_rate)]
segment_2_audio = audio_signal[int(seg2_start * sampling_rate):int(seg2_end * sampling_rate)]
abrupt_segment_audio = audio_signal[int(abrupt_seg_start * sampling_rate):int(abrupt_seg_end * sampling_rate)]

print("Playing Segment 1:")
display(Audio(segment_1_audio, rate=sampling_rate))

print("Playing Segment 2:")
display(Audio(segment_2_audio, rate=sampling_rate))

print("Playing Abruptly Segmented Speech:")
display(Audio(abrupt_segment_audio, rate=sampling_rate))
```
